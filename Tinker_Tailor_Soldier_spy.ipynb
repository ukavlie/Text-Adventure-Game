```python
"""
A) INTRODUCTION

This game is extremely loosely based on the novel Tinker ,Tailor, Soldier, Spy.
So loosely indeed, that the only common thing is the character names, its genre 
and its time period.

The game consist of four stages, and also defined functions for start (def SPY(),
fail and hallway (connecting point between the stages)
The objective is to collect evidence and use abductive reasoning 
to figure out who is the mole.

Stage A: Allelines Room
Stage B: Esterhases Room
Stage C: Haydons Room
Stage D: Final stage where you tell your theory 

B) KNOWN BUGS AND/OR ERRORS
None

"""
#from sys import exit
import random
import time
import sys
#list of suspects
suspects=["percy","toby","bill"]

def SPY():
    #picks a random person from the list to give game replay value
    global mole
    mole=random.choice(suspects)
    #introduction part of the game
    global name
    #global name as it stays the same throughout the game
    name=input(prompt="""
Major Westerby:
Welcome, you must be the intelligence officer that MI6 sent us.
What is your name, sir?
""")
    print(f"""
Nice to meet you, {name}! As you probably have been briefed about, our 
battalion have for the last weeks been leaking like a sinking ship.
It seems that we have a mole among our officers!""")
    #sleep to give the game a better flow
    #introduction to the suspects
    time.sleep(5)
    print("""
Three days ago, on October 6th, one of my officers handed over 
classified information to a Soviet spy. Luckily, we were able to 
intercept the spy, but he wont give us any information on the mole.
""")
    input(prompt="Press enter to continue")
    print("""
We have narrowed it down to three suspects.
""")
    time.sleep(1.2)
    print("\t1st Lieutenant Alleline\n")
    time.sleep(1.2)
    print("\t1st Lieutenant Haydon\n")
    time.sleep(1.2)
    print("\t2nd Lieutenant Esterhase\n")
    time.sleep(1.2)
    print("""
None of these officers were in the camp that day.
""")

    print("""
Your mission is to investigate each of the officers room. 
Find out where they were October the 6th and who's the mole. 
While we walk to the barracks 
you should read the briefs we have on each of them.

    1) Alleline, Percy
    2) Esterhase, Toby
    3) Haydon, Bill
""")

#three following variables are the briefs that will be printed in accordance to user input
#gives a cleaner code further down in my opinion
    Alleline="""
Name: Alleline, Percy
Birth: April 28th, 1933
Rank: 1st Lieutenant
Background:
Raised in an orphanage in East London. Received scholarship from 
London School of Economics and graduated with a degree in social science. 
Cultural refined for his background and enjoys the occasional concert 
with London Philharmonic Orchestra. 
"""
    Esterhase="""
Name: Esterhase, Toby
Birth: September 3rd, 1935
Rank: 2nd Lieutenant
Background
Born into nobility in Oxford and took his education in economics at the 
local university. Keeps to himself and has no connections outside his 
hometown or work. No hobbies or interest either. 
"""
    Haydon="""
Name: Haydon, Bill
Birth: February 10th, 1936
Rank: 1st Lieutenant
Born to an upper middle-class family of doctors. Went a different path
and graduated with a degree in Modern Languages from Cambridge.
An avid soccer fan devoted to Manchester United, watch their every game.
"""
    brief_lst=[Alleline,Esterhase,Haydon]
    brief_count=3
    #making brief into list so that one while loop is sufficient
    #Extra while loop to pick up human error
    brief=input(prompt="""
Which brief would you like to read?
    """)
    while brief_count>0:
        while brief_count>0:
            if "1" in brief or "alleline" in brief or "percy" in brief:
                brief_index=0
                brief_count-=1
                break
            elif "2" in brief or "esterhase" in brief or "toby" in brief:
                brief_index=1
                brief_count-=1
                break
            elif "3" in brief or "haydon" in brief or "bill" in brief:
                brief_index=2
                brief_count-=1
                break
            else:
                brief=str.lower(input(prompt="""
Which brief did you say you wanted to read again?
I didn't quite catch that.
    """))
                brief_index=None
                brief_count-=1
                continue
        try:
            story = brief_lst[brief_index]
            for char in story:
                sys.stdout.write(char)
                sys.stdout.flush()
                time.sleep(0.03)
            if brief_count==0:
                break
            else:
                brief=str.lower(input(prompt="""
Which brief would you like to read next?
"""))

        
        except:
            break

    print("""

Major Westerby:
We're here, so thats all the time we have. 
All three of them are out excercising with their soldiers, 
but not for much longer. Get moving to their rooms.
Time is of the essence, so don't use too much time in each room.
AND DON'T GET CAUGHT!
    """)
    #input to let the player breathe and take a break before movng to hallway
    input(prompt="Press enter to continue")
    print("""
You stand in the hallway of the officer barracks. Each of their rooms
are in front of you.""")
    time.sleep(1)
    hallway()
    
    
#the hallway were player can choose which room to investigate
def hallway():
        room_search=str.lower(input(prompt="""
What would you like to do?
    
        1) Search Allelines room
        2) Search Esterhases room
        3) Search Haydons room
        
        5) Report back to Major Westerby
        
"""))
#lets user decide path and sends the user to the different rooms
        if "1" in room_search or "alleline" in room_search:
            alleline()
        elif "2" in room_search or "esterhase" in room_search:
            esterhase()
        elif "3" in room_search or "haydon"  in room_search:
            haydon()
        elif "5" in room_search or "westerby" in room_search or "major" in room_search:
            report()
#controls for mistyping or bugs
        else:
            print("""
That's not your mission""")
            hallway()
            
            
            
            
            
def alleline():
#as with suspects the following list
#and random command ensures replay value
    furniture=["desk","closet","nightstand"]
    hiding_place=random.choice(furniture)
#different output dependent of who is the mole
    if mole=="percy":
        #raised in orphanage, no father to attend a birthday to
        daily_planner="At October 6th Lieutenant was attending his fathers birthday"
    else:
        daily_planner="At October 6th Lieutenant was attending a concert with\nLondon Philharmonic Orchestra"


    print("""
Culturally refined indeed, you think to yourself as you 
walk into Lieutenant Allelines room. Modern art is hanging on every wall
and in the corner a clarinet and a music stand with what looks like 
sheets from a classical concerto.""")

    print("""
This is however not the time to admire culture, you came here for a mission
        """)

    print("""
You notice three things that should be considered investigating more closely.""")
    search=random.randint(3,5)
#random limit to how much of the room can be investigated
    while search>0:
        invest=str.lower(input(prompt="""
What would you like to do?
            1) Investigate desk
            2) Investigate military closet
            3) Investigate nightstand
            
            5) Get out
"""))
#lets the player decide the order of his/her investigation
        
        
        
        #code for searching desk
        if "1" in invest or "desk" in invest:
            search-=1

            print("""
You walk towards the desk. The table itself is clean. 
However, there are two drawers.""")
            drawer=2
#while loop that breaks the loop after both drawers have been opened
            while drawer>0:
                desk=str.lower(input(prompt="""
Chose which drawer you would like to open, 
or press any other button to look other places?

        1 or 2

"""))
                if "1" in desk:
                    
                    #different outputs in regard to whether or not this is the hiding place
                    if hiding_place=="desk":
                        print(f"""
A daily planner, lets see what Alleline was doing on October 6th.

{daily_planner}""")
                    else:
                        print("""
It's empty""")
                    time.sleep(1)
                    #subtract from the overall "search-window"
                    search-=0.5
                    drawer-=1
                elif "2" in desk:
                    print("""
Some more music sheets""")
                    time.sleep(1)
                    search-=0.5
                    drawer-=1

                else: 
                    break
            continue
            
            
            
        #code foor investigating the closet
        elif "closet" in invest or "2" in invest:
            search -=1

            if hiding_place=="closet":
                print(f"""
A daily planner, lets see what Alleline was doing on October 6th.
{daily_planner}""")
                continue

            else:
                print("""
Nothing of interest here, only neatly folded army clothes""")
                continue

                
                
                
        #code for investigating nightstand        
        elif "3" in invest or "night" in invest:
            search-=1
            if hiding_place=="nightstand":
                print(f"""
A daily planner is lying at the nightstand, lets see what Alleline was doing on October 6th.

{daily_planner}""")
            else:
                print("The nightstand is empty, nothing in the drawer either.")
        else:
            #this will lead the player to walk out of Allelines room
            print("""
You walk out of Allelines room again and into the hallway.""")
            hallway()
    
    
    #reminds the player that the time stayed in each room is limited
    time.sleep(2)
    print("""
You hear steps from the hallway, time to get out!""")
    #player can either stay and get caught or get out
    action=str.lower(input(prompt="""
     
    1) Get out
    2) Stay for a little more and investigate
    
"""))
    if "1" in action or "out" in action:
        hallway()
    else:
        print("""
        
WHO ON EARTH ARE YOU AND WHAT THE HELL ARE YOU DOING IN MY ROOM!?!?""")
        time.sleep(1)
        print("""
You turn around and see Percy Alleline standing in the door.

You have been compromised""")
        time.sleep(2)
        fail()
        
        
        
        
        
def esterhase():
    #same as in every room, hiding place randomly picked
    furniture=["trash","closet"]
    hiding_place=random.choice(furniture)
    if mole=="toby":
        #No friend or hobbies only family n Oxford
        #therefor no reason to travel to Cambridge
        ticket="It's a return ticket from Cambridge dating October 6th"
    else:
        ticket="It's a return ticket from Oxford dating October 6th"


    print("""
It strikes you at ones that this is a person without any interests.
The room is spartan, to the border of sad.""")

    print("""
Besides a neatly made bed there are only the standard military closet
and a trash can""")
    search=3.5
    while search>0:
        invest=str.lower(input(prompt="""
What would you like to do?
        1) Investigate the trash can
        2) Investigate the military closet

        5) Get out

    """))
        
        
        
        #code for looking through the trash
        if "1" in invest or "trash" in invest:
            search-=1

            print("""
The trash has clearly not been taken out for a while.""")
            if hiding_place=="trash":
                print(f"""
On top of the pile you see what looks like a train ticket
You pick it up to look closer. 
{ticket}""")
                continue
            else:
                print("""
You glimpse over it, and wonder whats more depressing, this trash or the
decor of this room.""")
                continue    

                
                
                
                
         #code for looking through the close       
        elif "closet" in invest or "2" in invest:
            search -=1
            print("""
A combination lock with four digits. 
Maybe try a combination related to his birthday?""")
            combination=1938
            attempts=3
            try_comb=input(prompt="""
Try a combination or press any button to continue search.

    """)
            
            #code for guessing code
            #contains while and try to limit number of attempts and correct for human error
            while attempts>0:
                #tries to convert input from str to int
                #if not possible, player will be kicked out of closet code
                try:
                    int_try_comb=int(try_comb)
                    if int_try_comb==combination:
                        print("The lock opens, lets see what if there is anything of interest")
                        if hiding_place=="closet":
                            print(f"""
A train ticket, lets see where Esterhase has been travelling.
{ticket}""")
                            break
                        else:
                            print("""
Nothing of interest here, only neatly folded army clothes""")
                            break
                    elif int_try_comb!=combination:
                        search-=0.5
                        attempts-=1
                        if attempts>0:
                            print("""
Wrong code, it doesn't open""")
                            try_comb=input(prompt="""
Try another combination or press enter to continues search.

""")
                            continue
                        else:
                            print("""
I dont have time for this, you say to yourself.""")
                            break
                        continue
                except:
                    print("Maybe search other things")
                    break
        #code for walking back to hallway
        else:
            print("""
You walk out of Esterhases room, feeling more depressed than when you entered
a moment ago.""")
            hallway()
    #reminds the player about the limited time in each room
    time.sleep(2)
    print("""
You feel it on you that it is time to get out""")
    action=str.lower(input(prompt="""

    1) Get out
    2) Stay for a little more and investigate
    """))
    #same code as in previous room
    if "1" in action or "out" in action:
        hallway()
    else:
        print("""
WHO ON EARTH ARE YOU AND WHAT THE HELL ARE YOU DOING IN MY ROOM!?!?""")
        time.sleep(1)
        print("""
You turn around and see Toby Esterhase standing in the door.

You have been compromised""")
        time.sleep(2)
        fail()

    
    
    
    
    
def haydon():
    furniture=["trash","nightstand","table"]
    hiding_place=random.choice(furniture)
    if mole=="bill":
        #A fan of Manchester United would never attend a match consisting of only thier enemies
        ticket="It's a football ticket for the match between Manchester City and Liverpool"
    else:
        ticket="It's a football ticket for the match between Manchester United and Liverpool"


    print("""
This is a fan of Manchester United, you think to yourself. 
A poster of the team is hanging over his bed, and pictures of the
different players pride one of the walls""")

    print("""
There are however three things you find interesting to your investigation""")
    search=random.randint(3,5)
    while search>0:
        invest=str.lower(input(prompt="""
What would you like to do?
            1) Investigate the trash can
            2) Investigate the nightstand
            3) Investigate the table

            5) Get out

"""))

        if "1" in invest or "trash" in invest:
            search-=1

            if hiding_place=="trash":
                print(f"""
It's been emptied recently, but there is one item at the bottom.
You pick it up to look closer.

{ticket}""")
                continue
            else:
                print("""
It's been emptied recently!""")
                continue

        elif "2" in invest or "night" in invest:
            search-=1
            if hiding_place=="nightstand":
                print(f"""
A ticket is lying at the nightstand, lets see what Haydon was doing on October 6th.

{ticket}""")
            else:
                print("The nightstand is empty, nothing in the drawer either.")

        elif "3" in invest or "desk" in invest:
            search-=1

            print("""
You walk towards the table. It has clearly been used as a desk
with papers floating around.""")

            table=str.lower(input(prompt="""
What would you like to do?

        1) Search through the papers
        2) Get away from the mess and search the other things

"""))
            if "1" in table or "search" in table:
                if hiding_place=="table":
                    print(f"""
Among the papers you find a ticket.

    {ticket}""")
                else:
                    print("""
Just a lot of mess. I thought officers were more tidy than this""")
                time.sleep(1)
                search-=0.5
                continue


            else:
                continue

        else:
            print("""
You return to the hallway.""")
            hallway()

    time.sleep(2)
    print("""
A loud noise from the outside makes you jump up.
Thats it for this room, time to get out""")
    action=input(prompt="""

    1) Get out
    2) Stay for a little more and investigate
    """)
    if "1" in action or "out" in action:
        hallway()
    else:
        print("""
WHO ON EARTH ARE YOU AND WHAT THE HELL ARE YOU DOING IN MY ROOM!?!?""")
        time.sleep(1)
        print("""
You turn around and see Bill Haydon standing in the door.

You have been compromised""")
        #if player overstays his visit, he will fail
        fail()
        
        
        
        
        
def fail():
    #transelates mole to a full name
    if mole=="toby":
        mole_full_name="Toby Esterhase"
    elif mole=="bill":
        mole_full_name="Bill Haydon"
    else:
        mole_full_name="Percy Alleline"
    
    print(f"""
        Your mission has failed!
        The mole was {mole_full_name}""")
    #replay function where player can chose to replay or quit
    replay=str.lower(input(prompt="""
Would you like to start over again?

    1) Yes
    2) No
"""))
    if "1" in replay or "y" in replay:
        SPY()
    else:
        print("Thanks for now")
        exit

        
        

def report():
    #function were the player can use information
    #gathered through the game to come to a conclusion
    report=str.lower(input(prompt=f"""
Major Westerby:
So {name}, do you know who the mole is?

    1) Yes
    2) No
    """))
    if "2" in report or "no" in report:
        print("Thats too bad")
        fail()
        #if player hasn't figured out who the mole is, the player fails
    else:
        suspect=str.lower(input(prompt="""
Well, who is it?

    1) Alleline, Percy
    2) Esterhase, Toby
    3) Haydon, Bill
    """))
        #player can state his suspect
        while True:
            if "1" in suspect or "alleline" in suspect or "percy" in suspect:
                suspect="percy"
                break
            elif "2" in suspect or "esterhase" in suspect or "toby" in suspect:
                suspect="toby"
                break
            elif "3" in suspect or "haydon" in suspect or "bill"in suspect:
                suspect="bill"
                break
            else:
                suspect=str.lower(input(prompt="""
I'm sorry, I didn't catch that
You will have to repeat the name:

    1) Alleline, Percy
    2) Esterhase, Toby
    3) Haydon, Bill
    """))
                continue
                #while loop to catch up human error
        # if the players suspect match the games mole, the player win!
        if suspect==mole:
            print(f"""
{"*"*80}

                              CONGRATULATIONS

                            YOU CAUGTH THE MOLE
    
{"*"*80}""")
            time.sleep(5)
            replay=str.lower(input(prompt="""
Would you like to start over again?

    1) Yes
    2) No
"""))
            if "1" in replay or "y" in replay:
                SPY()
            else:
                print("Thanks for now")
                exit
        else:
            print("That is not the right mole")
            fail()

SPY()
    
```
